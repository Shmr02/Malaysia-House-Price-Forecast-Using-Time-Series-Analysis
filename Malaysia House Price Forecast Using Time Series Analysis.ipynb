# **DOUBLE EXPONENTIAL SMOOTHING**

```python
data.columns
```

```python
data['date'] = pd.to_datetime(data['date'], format='%Y/%m/%d')
```

```python
# Split the data into in-sample (01/10/2023 to 30/09/2024) and out-sample (01/10/2024 onwards)
in_sample = data[(data['date'] >= '2009-3-31') & (data['date'] <= '2020-12-31')]
out_sample = data[data['date'] > '2020-12-31']
```

```python
in_sample.head()
```

```python
HP_in_sample = in_sample['value']
HP_out_sample = out_sample['value']
```

```python
HP_out_sample.count()
```

```python
HP_in_sample.head()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Calculate initial level and trend
# Access the first value of the 'house_price' column
initial_level = in_sample['value'].iloc[0]
initial_trend = 0

# Fit the Double Exponential Smoothing model with known initialization
# Pass only the 'house_price' column to ExponentialSmoothing
fit1 = ExponentialSmoothing(in_sample['value'], trend='add', initialization_method="known", initial_level=initial_level, initial_trend=initial_trend).fit(smoothing_level=0.2, smoothing_trend=0.2, optimized=False)
fcast1_in_sample = fit1.fittedvalues
fcast1_out_sample = fit1.forecast(len(out_sample)).rename(r"$\alpha=0.2, \beta=0.2$")
```

```python
fcast1_in_sample.head()
```

```python
fcast1_out_sample.head()
```

```python
fit2 = ExponentialSmoothing(HP_in_sample, trend='add', initialization_method="heuristic").fit(smoothing_level=0.6, smoothing_trend=0.6, optimized=False)
fcast2_in_sample = fit2.fittedvalues
fcast2_out_sample = fit2.forecast(len(HP_out_sample)).rename(r"$\alpha=0.6, \beta=0.6$")
```

```python
fit3 = ExponentialSmoothing(HP_in_sample, trend='add', initialization_method="estimated").fit()
fcast3_in_sample = fit3.fittedvalues
fcast3_out_sample = fit3.forecast(len(HP_out_sample)).rename(r"$\alpha=%s, \beta=%s$" % (fit3.model.params["smoothing_level"], fit3.model.params["smoothing_trend"]))
print(fit3.summary())
```

```python
fcast3_in_sample
```

```python
fcast3_out_sample
```

```python
import matplotlib.pyplot as plt

# Create the plot
plt.figure(figsize=(12, 8))

# Plot actual in-sample data
plt.plot(in_sample['date'], in_sample['value'], marker="o", color="black", label='Actual In-Sample')

# Plot actual out-of-sample data
plt.plot(out_sample['date'], out_sample['value'], marker="o", color="gray", label='Actual Out-Sample')

# Plot in-sample forecasts
plt.plot(in_sample['date'], fcast3_in_sample, marker="o", color="blue", label='Forecast In-Sample')

# Plot out-sample forecasts
plt.plot(out_sample['date'], fcast3_out_sample, marker="o", color="blue", linestyle="--", label='Forecast Out-Sample')

# Add legend
plt.legend()

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Actual vs Forecast exponential smoothing model')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Plot the results
plt.figure(figsize=(12, 8))

plt.plot(in_sample['date'], fcast3_in_sample, color="green")
(line3,) = plt.plot(out_sample['date'], fcast3_out_sample, color="green")
plt.legend([ line3], [ fcast3_out_sample.name])
plt.xlabel('Date')
plt.ylabel('House Price')
plt.title('Double Exponential Smoothing')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error

# Calculate MSE, MAE, and MAPE for each model (in-sample)
mse_fit1_in_sample = mean_squared_error(HP_in_sample, fcast1_in_sample)
mae_fit1_in_sample = mean_absolute_error(HP_in_sample, fcast1_in_sample)
mape_fit1_in_sample = mean_absolute_percentage_error(HP_in_sample, fcast1_in_sample)* 100

mse_fit2_in_sample = mean_squared_error(HP_in_sample, fcast2_in_sample)
mae_fit2_in_sample = mean_absolute_error(HP_in_sample, fcast2_in_sample)
mape_fit2_in_sample = mean_absolute_percentage_error(HP_in_sample, fcast2_in_sample) * 100

mse_fit3_in_sample = mean_squared_error(HP_in_sample, fcast3_in_sample)
mae_fit3_in_sample = mean_absolute_error(HP_in_sample, fcast3_in_sample)
mape_fit3_in_sample = mean_absolute_percentage_error(HP_in_sample, fcast3_in_sample) * 100



print(f"Mean Squared Error (In-Sample Alpha=0.2): {mse_fit1_in_sample:.2f}")
print(f"Mean Absolute Error (In-Sample Alpha=0.2): {mae_fit1_in_sample:.2f}")
print(f"Mean Absolute Percentage Error (In-Sample Alpha=0.2): {mape_fit1_in_sample:.2f}%")

print(f"\nMean Squared Error (In-Sample Alpha=0.6): {mse_fit2_in_sample:.2f}")
print(f"Mean Absolute Error (In-Sample Alpha=0.6): {mae_fit2_in_sample:.2f}")
print(f"Mean Absolute Percentage Error (In-Sample Alpha=0.6): {mape_fit2_in_sample:.2f}%")

print(f"\nMean Squared Error (In-Sample Optimal Alpha): {mse_fit3_in_sample:.2f}")
print(f"Mean Absolute Error (In-Sample Optimal Alpha): {mae_fit3_in_sample:.2f}")
print(f"Mean Absolute Percentage Error (In-Sample Optimal Alpha): {mape_fit3_in_sample:.2f}%")
```

```python
# Calculate MSE, MAE, and MAPE for each model (out-sample)
mse_fit1_out_sample = mean_squared_error(HP_out_sample, fcast1_out_sample)
mae_fit1_out_sample = mean_absolute_error(HP_out_sample, fcast1_out_sample)
mape_fit1_out_sample = mean_absolute_percentage_error(HP_out_sample, fcast1_out_sample) * 100

mse_fit2_out_sample = mean_squared_error(HP_out_sample, fcast2_out_sample)
mae_fit2_out_sample = mean_absolute_error(HP_out_sample, fcast2_out_sample)
mape_fit2_out_sample = mean_absolute_percentage_error(HP_out_sample, fcast2_out_sample) * 100

mse_fit3_out_sample = mean_squared_error(HP_out_sample, fcast3_out_sample)
mae_fit3_out_sample = mean_absolute_error(HP_out_sample, fcast3_out_sample)
mape_fit3_out_sample = mean_absolute_percentage_error(HP_out_sample, fcast3_out_sample) * 100


print(f"Mean Squared Error (Out-Sample Alpha=0.2): {mse_fit1_out_sample:.2f}")
print(f"Mean Absolute Error (Out-Sample Alpha=0.2): {mae_fit1_out_sample:.2f}")
print(f"Mean Absolute Percentage Error (Out-Sample Alpha=0.2): {mape_fit1_out_sample:.2f}%")

print(f"\nMean Squared Error (Out-Sample Alpha=0.6): {mse_fit2_out_sample:.2f}")
print(f"Mean Absolute Error (Out-Sample Alpha=0.6): {mae_fit2_out_sample:.2f}")
print(f"Mean Absolute Percentage Error (Out-Sample Alpha=0.6): {mape_fit2_out_sample:.2f}%")

print(f"\nMean Squared Error (Out-Sample Optimal Alpha): {mse_fit3_out_sample:.2f}")
print(f"Mean Absolute Error (Out-Sample Optimal Alpha): {mae_fit3_out_sample:.2f}")
print(f"Mean Absolute Percentage Error (Out-Sample Optimal Alpha): {mape_fit3_out_sample:.2f}%")
```
